```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Bayern Munich").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(5, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling.to_csv("Bundesliga rolling.csv")
```

```python
from sklearn.metrics import precision_score
```

```python
#Random Forest
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2018-01-01']
    test = data[data["date"] > '2018-01-01']
    rf.fit(train[predictors], train["target"])
    predrf = rf.predict(test[predictors])
    combinedrf = pd.DataFrame(dict(actual=test["target"], predicted=predrf), index=test.index)
    RFprecision= precision_score(test["target"], predrf)
    return combinedrf, RFprecision
```

```python
combinedrf, RFprecision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
print ("Precision of Random Forest", RFprecision)
```

```python
print ("Accuracy of Random Forest",accuracy_score(test["target"], predrf))
```

```python
from sklearn import metrics
```

```python
mse = round(metrics.mean_squared_error(test["target"], predrf), 4)
print ("Mean Squared Error of RF",mse)
```

```python
from sklearn.metrics import classification_report
print('Classification Report of Random Forest: \n',classification_report(test["target"], predrf,digits=4))
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
auc = roc_auc_score(test["target"], predrf)
```

```python
fpr, tpr, thresholds = roc_curve(test["target"], predrf)
plt.figure()
plt.plot(fpr, tpr, label='Random Forest (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic of Random Forest')
plt.legend(loc="lower right")
plt.savefig('RF_ROC')
plt.show()
```
